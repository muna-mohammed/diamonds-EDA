```python
data["x*y*z"] = data.apply(lambda row: row["x"] * row["y"] * row["z"], axis=1)
data.head()
# make a new column called x*y*z 
```

```python
# determine X, and y for spliting data and training model
X = data[["carat","cut", "color", "depth", "x*y*z", "table", "clarity"]]
y = data[['price']]
```

```python
X = pd.get_dummies(data=X)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
model = LinearRegression().fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
